```python
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
from sklearn import svm
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
import matplotlib.pyplot as plt

# Load the spam dataset
spam = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/cleaned_EmailSpamDataSet.csv')

# Split the dataset into training and testing sets
x_train, x_test, y_train, y_test = train_test_split(spam['Message'], spam['Category'], test_size=0.2, random_state=42)

# Convert the email text into numerical features using CountVectorizer
cv = CountVectorizer()
features_train = cv.fit_transform(x_train)
features_test = cv.transform(x_test)

# Train a Support Vector Machine (SVM) classifier
model = svm.SVC()
model.fit(features_train, y_train)

# Make predictions on the test set
predictions = model.predict(features_test)

# Calculate and print accuracy, precision, recall, and F1 score
accuracy = accuracy_score(y_test, predictions)
precision = precision_score(y_test, predictions, average='macro')
recall = recall_score(y_test, predictions, average='macro')
f1 = f1_score(y_test, predictions, average='macro')

print(f"Accuracy: {accuracy:.2f}")
print(f"Precision: {precision:.2f}")
print(f"Recall: {recall:.2f}")
print(f"F1 Score: {f1:.2f}")
```
